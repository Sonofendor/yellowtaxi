# Демонстрация проекта

```python
import pandas as pd
import numpy as np
from warnings import filterwarnings
from ipywidgets import widgets
from tools import * #модуль с функциями, использующимися для работы ноутбука
```

```python
filterwarnings('ignore')
data = pd.read_csv('taxi_trips.csv',index_col=0,parse_dates=[0])
data.head()
```

По-хорошему, для демонстрации данного проекта нужно создать сайт с интерактивной картой, чтобы все вычсилялось на сервере и карты с графиками отображались на любом компьюетере, даже без питона и уж тем более без библиотек. Однако у меня нет возможности и желания поднимать свой сервер 24/7, так что придется демонстрировать в ноутбуке.

Здесь представлена демонстрация двух видов: **бизнес-демонстрация** - это демонстрация конечного продукта для заказчика. А именно - интерактивная карта, на которой можно выбрать дату и время и увидеть предсказанные значения количества поездок в каждом районе. **Техническая демонстрация** - это скорее для аналитиков со стороны бизнеса, чтобы показать им, что модель работает вполне адекватно и предъявить качество.

## Бизнес-демонстрация

```python
def display_prediction(datetime=pd.to_datetime('2016-05-09 12:00:00')):
    return embed_map(create_map(multiprediction(data,datetime)))
```

**Инструкция**: необходимо выбрать время из представленного списка. На карте будет отображена сетка с цветовым окрасом в зависимости от количества поездок в выбранный час. Точное значение количества поездок можно узнать, нажав на интересующую вас ячейку. *Построение прогнозов может занять некоторое время.*

```python
widgets.interact(display_prediction, datetime=pd.date_range(start='2016-01-01 00:00:00',end='2016-06-30 23:00:00',freq='H'))
```

## Техническая демонстрация

```python
def plots(area, start=pd.to_datetime('2016-05-01 00:00:00'), stop=pd.to_datetime('2016-05-31 23:00:00')):
    plot_prediction(one_area_prediction(data, area, start, stop))
```

Здесь необходимо выбрать зону, интересующую вас (местоположение зоны можно узнать на карте выше), начало интересующего вас периода и его конец. На графике будут предсталвены ряд реальных значений количества поездок в данный период и ряд предсказаний модели. Также будет представлен график ошибок модели, средняя абсолютная, средняя квадратическая и средняя ошибки за данный период.

```python
daterange = pd.date_range(start='2016-01-01 00:00:00',end='2016-06-30 23:00:00',freq='H')
display(widgets.interact(plots, start=daterange,
                 stop=daterange,area=data.columns))
```
